```python
import numpy as np
from diffeobrot import diffeobrot_set
from PIL import Image 
from matplotlib import cm
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

## Continuous time Mandelbrot dynamical system with lag

$$
\frac{dz}{dt}(t) + z(t) = z(t-t_0)^2+c,\\
z(t) = 0 \forall t \leq 0,
$$
where $c$ is a complex number.

For small lag $t_0$, the discretization scheme for the above dynamics matches the discrete Mandelbrot equation. As the lag increases, the system stabilizes and the resulting Mandelbrot set gets "smoother".

```python
params = {
    'figure.figsize': (11,28),
}
plt.rcParams.update(params)
```

```python
width, height = 1920, 1080
xmin, xmax, ymin, ymax = -2.0,0.5,-1.25,1.25
maxiter = 518
dt = 1e-1
```

```python
t0 = 0.05

diffeobrot = diffeobrot_set(xmin, xmax, ymin, ymax, width, height, t0, dt, maxiter, method='parallel')
im = Image.fromarray(np.uint8(cm.gist_earth(diffeobrot)*255))
plt.axis('off')
plt.imshow(im)
```
