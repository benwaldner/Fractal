```python
import numpy as np
from julia import julia_set
from PIL import Image 
from matplotlib import cm
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
params = {
    'figure.figsize': (11,28),
}
plt.rcParams.update(params)
colormap = 'gist_earth'
```

```python
width, height = 1920, 1080
xmin, xmax, ymin, ymax = -2.0, 2.0, -2.0, 2.0
maxiter = 128

cs = [
    0.285 + 0.01j,
    -0.8j,
    -0.7269 + 0.1889j,
    -0.835 - 0.2321j,
    -0.4 + 0.6j,
]

#phi = 0.5*(1+np.sqrt(5))+0j
#c = 1-phi

for c in cs:
    julia = julia_set(xmin, xmax, ymin, ymax, width, height, maxiter,\
                              c=c, method='cuda')
    im = Image.fromarray(np.uint8(getattr(cm, colormap)(julia)*255))
    fig = plt.figure()
    plt.axis('off')
    plt.imshow(im)
    #im.save('julia_{}_{}.tif'.format(c.real, c.imag))
```
