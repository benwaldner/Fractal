```python
import numpy as np
from burning_ship import burning_ship_set
from PIL import Image 
from matplotlib import cm
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
params = {
    'figure.figsize': (11,28),
}
plt.rcParams.update(params)
```

```python
width, height = 1920, 1080
xmin, xmax, ymin, ymax = -2.25,1.25,-2.25,2.25
maxiter = 1024

burning_ship = burning_ship_set(xmin, xmax, ymin, ymax, width, height, maxiter, method='parallel')
```

```python
im = Image.fromarray(np.uint8(cm.nipy_spectral(burning_ship)*255))
```

```python
plt.axis('off')
plt.imshow(im)
```

```python
im.save('burning_ship.tif')
```
