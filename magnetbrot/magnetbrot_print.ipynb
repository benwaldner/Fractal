```python
import numpy as np
from magnetbrot import magnetbrot_set
from PIL import Image 
from matplotlib import cm
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
params = {
    'figure.figsize': (11,28),
}
plt.rcParams.update(params)
```

```python
width, height = 1920, 1080
xmin, xmax, ymin, ymax = -1.0,5.0,-2.5,2.5
maxiter = 512

magnetbrot = magnetbrot_set(xmin, xmax, ymin, ymax, width, height, maxiter, method='parallel')
```

```python
im = Image.fromarray(np.uint8(cm.gist_earth(magnetbrot)*255))
```

```python
plt.axis('off')
plt.imshow(im)
```

```python
im.save('magnetbrot.tif')
```
